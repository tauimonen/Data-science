```python
import pandas as pd
import numpy as np
```

```python
# Let's read some wether data
wh = pd.read_csv("https://raw.githubusercontent.com/csmastersUH/data_analysis_with_python_2020/master/kumpula-weather-2017.csv")
```

```python
wh.head()
```

```python
# Let's drop some comumns, as taking averages over these is not very interesting.
wh2 = wh.drop(["Year", "m", "d"], axis=1)  
wh2.mean()
```

```python
# Summary statistics for each numeric column as DataFrame.
wh.describe()
```

```python
wh2.describe()
```

```python
# Data from 2017
y2017 = wh[wh.Year == 2017]
y2017
```

```python
# Max value
m = y2017.loc[:, "Snow depth (cm)"].max()
m
```

```python
# Missing data

# Let’s use the unique method of the Series object to find out, 
# which different values are used in this column

wh["Snow depth (cm)"].unique()
```

```python
pd.Series([1,3,2]) # int64
```

```python
pd.Series([1,3,2, np.nan]) # float64
```

```python
pd.Series(["jack", "joe", None]) # missing value = object
```

```python
wh.isnull()
```

```python
# finding out all the rows that contain at least one missing value with any method
wh[wh.isnull().any(axis=1)]
```

```python
# The notnull method works conversively
wh[wh.notnull().any(axis=1)]
```

```python
# Drops rows, as default axis is 0
wh.dropna().shape
```

```python
# Drops the columns containing missing values
wh.dropna(axis=1).shape
```

```python
wh = wh.fillna(method='ffill')
wh[wh.isnull().any(axis=1)]
```
