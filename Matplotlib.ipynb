```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.array([2, 5, 7, 4, 7, 0, 3, 1, 9, 2])
```

```python
plt.plot(a)                  
plt.title("Figure") 
plt.xlabel("x-axis")       
plt.ylabel("y-axis")       
plt.show()    
```

```python
x1 = 2,4,6,7
y1 = 4,3,5,1
x2 = 1,2,3,4
y2 = 4,2,3,1
plt.title("title") 
plt.xlabel("x-axis")       
plt.ylabel("y-axis") 
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
# Subplots

fig, ax = plt.subplots(2,2)
print(ax.shape)
ax[0,0].plot(np.arange(10))          # top left
ax[0,1].plot(np.arange(6,0,-1))     # top right
ax[1,0].plot((-1)**np.arange(6))    # bottom left
ax[1,1].plot((-1)**np.arange(1,7))  # bottom right
plt.show()
```

```python
# 2 x 2 + index number

plt.subplot(2, 2, 1)    
plt.plot(np.arange(6))
plt.subplot(2, 2, 2)
plt.plot(np.arange(6, 0, -1))
plt.subplot(2, 2, 3)
plt.plot((-1)**np.arange(6))
plt.subplot(2, 2, 4)
plt.plot((-1)**np.arange(1, 7))
plt.show()
```
