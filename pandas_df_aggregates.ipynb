```python
import pandas as pd
import numpy as np
```

```python
# Read the data
wh = pd.read_csv("https://raw.githubusercontent.com/csmastersUH/data_analysis_with_python_2020/master/kumpula-weather-2017.csv")
wh.head()
```

```python
# change the column names
wh3 = wh.rename(columns={"m": "Month", "d": "Day", "Precipitation amount (mm)" : "Precipitation",
                         "Snow depth (cm)" : "Snow", "Air temperature (degC)" : "Temperature"})
wh3.head()
```

```python
# splits the DataFrame into groups
groups = wh3.groupby("Month")
groups
```

```python
# Number of groups (12 months)
len(groups)
```

```python
# Iterating through all the groups
for key, group in groups:
    print(key, len(group))
```

```python
# Group february
groups.get_group(2)  
```

```python
# a subset of columns
groups["Temperature"]
```

```python
# mean aggregation
groups["Temperature"].mean()
```

```python
groups["Precipitation"].sum()
```

```python
# -1.0 values in the Precipitation field are causing trouble -> convert to zeros
wh4 = wh3.copy()
wh4.loc[wh4.Precipitation == -1, "Precipitation"] = 0
wh4.loc[wh4.Snow == -1, "Snow"] = 0
wh4.head()
```

```python
wh4.groupby("Month")["Precipitation"].sum()
```

```python
# The aggregations are not the only possible operations on groups. The other possibilities 
# are filtering, transformation, and application.
```

```python
# # The filter function must return a boolean value
def myfilter(df):
    return df["Precipitation"].sum() >= 150

# Filter out months with total precipitation less that 150 mm
wh4.groupby("Month").filter(myfilter)
```

```python
# deviations from the monthly means by transformation
pd.concat([wh4.iloc[:, 0:3],
           wh4.groupby("Month")[["Precipitation", "Snow", "Temperature"]].transform(lambda x : x - x.mean())],
          axis=1)
```
