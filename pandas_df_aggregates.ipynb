```python
import pandas as pd
import numpy as np
```

```python
# Read the data
wh = pd.read_csv("https://raw.githubusercontent.com/csmastersUH/data_analysis_with_python_2020/master/kumpula-weather-2017.csv")
wh.head()
```

```python
# change the column names
wh3 = wh.rename(columns={"m": "Month", "d": "Day", "Precipitation amount (mm)" : "Precipitation",
                         "Snow depth (cm)" : "Snow", "Air temperature (degC)" : "Temperature"})
wh3.head()
```

```python
# splits the DataFrame into groups
groups = wh3.groupby("Month")
groups
```

```python
# Number of groups (12 months)
len(groups)
```

```python
# Iterating through all the groups
for key, group in groups:
    print(key, len(group))
```

```python
# Group february
groups.get_group(2)  
```

```python
# a subset of columns
groups["Temperature"]
```

```python
# mean aggregation
groups["Temperature"].mean()
```
