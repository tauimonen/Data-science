```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Loading a dataset

tips = sns.load_dataset('tips')
tips.head()
```

```python
# Creating two dataframes by grouping

tips_bill = tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].sum()
tips_tip = tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].sum()
```

```python
tips_bill
```

```python
# then we make two dataframes to train some merging

del tips_bill['tip']
```

```python
tips_bill
```

```python
del tips_tip['total_bill']
tips_tip
```

```python
# merging on the indexes
pd.merge(tips_bill, tips_tip, right_index=True, left_index=True)
```

```python
# reseting indexes and then merge on the columns
pd.merge(
    tips_bill.reset_index(), 
    tips_tip.reset_index(),
    on=['sex', 'smoker']
)
```

```python
# Also like this, but be carefull!

pd.merge(
    tips_bill.reset_index(), 
    tips_tip.reset_index()
)
```

```python
# partial column and index
pd.merge(
    tips_bill.reset_index(), 
    tips_tip,
    left_on=['sex', 'smoker'],
    right_index=True
)
```

```python
# combinations
tips_bill_s = tips_bill.reset_index(level=1)
tips_bill_s
```

```python
pd.merge(
    tips_tip.reset_index(), 
    tips_bill_s,
    on=['sex', 'smoker']
)
```

```python
# left join
pd.merge(
    tips_bill.reset_index(), 
    tips_tip.reset_index().head(2),
    how='left'
)
```

```python
# inner join
pd.merge(
    tips_bill.reset_index(), 
    tips_tip.reset_index().head(2),
    how='inner'
)
```

```python
# outer join + indicator
pd.merge(
    tips_bill.reset_index().tail(3), 
    tips_tip.reset_index().head(3),
    how='outer',
    indicator=True
)
```

```python
# columns with the same name
pd.merge(tips_bill, 
         tips_bill, 
         right_index=True, 
         left_index=True,
         suffixes=('_left', '_right')
)
```

```python
# row wise
pd.concat([tips_bill, tips_bill, tips_tip], sort=False)
```

```python
# column wise
pd.concat([tips_bill, tips_tip], axis=1)
```

```python
# adding keys
pd.concat([tips_bill, tips_tip], sort=False, keys=['num0', 'num1'])
```
